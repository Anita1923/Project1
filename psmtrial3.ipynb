```python
import os
import pandas as pd
from transformers import pipeline
from TTS.api import TTS
import gradio as gr

# Load pre-trained emotion detection model
emotion_classifier = pipeline("text-classification", model="j-hartmann/emotion-english-distilroberta-base", return_all_scores=True)

# Detect emotion in text
def detect_emotion(text):
    scores = emotion_classifier(text)
    emotion = max(scores[0], key=lambda x: x['score'])['label']
    return emotion

# Example usage
sample_text = "I'm so glad with the results!"
emotion = detect_emotion(sample_text)
print(f"Detected emotion: {emotion}")
```

```python
from TTS.api import TTS

# Initialize TTS model
tts_model = TTS(model_name="tts_models/en/vctk/vits", progress_bar=False, gpu=False)

# Function: Text-to-Speech with emotion and speaker options
def text_to_speech_with_emotion(text, emotion, speaker, output_path):
    # Adjust SSML for emotions
    if emotion == "sadness":
        ssml = f'<prosody pitch="-5%" rate="90%"><break time="300ms" strength="weak"/>{text}</prosody>'
    elif emotion == "anger":
        ssml = f'<prosody pitch="+10%" rate="110%"><emphasis level="strong">{text}</emphasis></prosody>'
    elif emotion == "joy":
        ssml = f'<prosody pitch="+5%" rate="110%">{text}</prosody>'
    else:
        ssml = text  # Neutral

    # Generate speech
    tts_model.tts_to_file(text=ssml, speaker=speaker, file_path=output_path)

```

```python
# Workflow: End-to-end emotion-aware TTS
def generate_emotion_aware_speech(text, speaker="p226"):
    emotion = detect_emotion(text)  # Detect emotion
    output_path = "emotion_speech_output.mp3"
    text_to_speech_with_emotion(text, emotion, speaker, output_path)  # Generate speech
    return f"Detected Emotion: {emotion}", output_path
```

```python
# Interactive Web Interface with Gradio
def tts_interface(text, speaker):
    emotion, audio_path = generate_emotion_aware_speech(text, speaker)
    return emotion, audio_path, audio_path

# Define Gradio interface
interface = gr.Interface(
    fn=tts_interface,
    inputs=[
        gr.Textbox(lines=5, label="Input Text"),
        gr.Dropdown(["p226", "p227", "p228"], label="Select Speaker"),
    ],
    outputs=[
        gr.Textbox(label="Detected Emotion"),
        gr.Audio(label="Emotion-Aware Speech"),
        gr.File(label="Download Audio File"),
    ],
    title="Emotion-Aware TTS for Smart Library",
    description="Enter text and select a speaker to generate emotion-aware speech for Smart Library applications."
)
```

```python
# Launch interface
interface.launch()
```
